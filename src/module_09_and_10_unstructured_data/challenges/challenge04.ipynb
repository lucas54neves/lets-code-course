### 1 - Trocar cv2.getRotation por uma função com numpy

```python
def my_get_rotation(image, degrees=90):
    if degrees == 90:
        _image = np.rot90(image)
    elif degrees == 180:
        _image = np.rot90(image, 2)
    elif degrees == 270:
        _image = np.rot90(image, 3)
    else:
        _image = image
    return _image
```

```python
for (image_name, image) in images:
    plt.figure(figsize = (20, 6))

    plt.subplot(1,4,1)
    plt.title(f"{image_name}")
    plt.imshow(image)
    plt.box(False)

    plt.subplot(1,4,2)
    plt.title(f"{image_name} 90º")
    _image = my_get_rotation(image, 90)
    plt.imshow(_image)
    plt.box(False)

    plt.subplot(1,4,3)
    plt.title(f"{image_name} 180º")
    _image = my_get_rotation(image, 180)
    plt.imshow(_image)
    plt.box(False)

    plt.subplot(1,4,4)
    plt.title(f"{image_name} 270º")
    _image = my_get_rotation(image, 270)
    plt.imshow(_image)
    plt.box(False)
```

### 2 - Mudança para escala cinza utilizando array(numpy)

```python
plt.figure(figsize = (20, 6))

for index, (image_name, image) in enumerate(images):
    gray = image.copy()

    red = gray[:, :, 0] * 0.299
    green = gray[:, :, 1] * 0.587
    blue = gray[:, :, 2] * 0.114

    matrix = red + green + blue

    plt.subplot(1,3,index+1)
    plt.title(f"{image_name} gray")
    plt.imshow(matrix, cmap = 'gray')
    plt.box(False)
```
