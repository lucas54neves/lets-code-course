# Questão 1
Faça um programa que peça 10 números para o usuário e guarde os números em uma lista. Com a lista preenchida descubra quantos números são pares.

````python
numbers = []

count = 0

while count < 10:
  number = int(input('Insira um número: '))

  numbers.append(number)

  count += 1

count = 0

pair_numbers = 0

while count < 10:
  if numbers[count] % 2 == 0:
    pair_numbers += 1

  count += 1

print('Quantidade de numeros pares: {}'.format(pair_numbers))
````

# Questão 2
Faça um script que leia 10 números do usuário e devolva duas listas: uma com os números ímpares e outra com os números pares.
Exemplo:
```
input 1: 5
input 2: 4
(etc...): 6.. 7.. 1.. 2.. 3.. 7..
input 9: 4
input 10: 8
output: lista com ímpares: [5, 7, 1, 3, 7]
        lista com pares: [4, 6, 2, 4, 8]
```

````python
count = 0

pair_numbers = []

odd_numbers = []

while count < 10:
  number = int(input('Entre com o numero: '))

  if number % 2 == 0:
    pair_numbers.append(number)
  else:
    odd_numbers.append(number)

  count += 1

print('Lista com impares: {}'.format(odd_numbers))

print('Lista com pares: {}'.format(pair_numbers))
````

# Questão 3

Faça um script que leia números do usuário, enquanto ele não digitar 0. Armazene esses números em uma lista e ao final informe quantos números foram digitados, ignorando o 0.

Exemplo:

```
input: 2
input: 4
input: -1
input: -5
input: 0
output: Você digitou 4 números
```

````python
number = int(input('Entre com um numero: '))

numbers = []

while number != 0:
  numbers.append(number)

  number = int(input('Entre com um numero: '))

print('Voce digitou {} numeros'.format(len(numbers)))
````

# Questão 4

Faça um script que lê 10 números e mostra na tela aqueles que são maiores que a média.

````python
count = 0

numbers = []

sum = 0

while count < 10:
  number = float(input('Entre com um numero: '))

  numbers.append(number)

  sum += number

  count += 1

average = sum / 10

count = 0

largers_than_average = []

while count < 10:
  if numbers[count] > average:
    largers_than_average.append(numbers[count])
  
  count += 1

print(largers_than_average)
````

# Questão 5

Crie duas listas sobre um time de futebol com 11 jogadores: uma lista com a idade dos jogadores e outra com as alturas. Mostre na tela a idade do jogador mais alto.

````python
ages = [20,60,42,1,2,35,4,85,84,52,32,12]

heights = [1.5,1.6,1.8,1.8,1.9,1.8,2.3,1.9,1.3,1.8,1.3]

tallest = [ heights[0], ages[0] ]

count = 1

while count < 10:
  if tallest[0] < heights[count]:
    tallest = [ heights[count], ages[count] ]
  
  count += 1

print(tallest[1])
````
