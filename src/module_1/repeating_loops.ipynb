# Questão 1
Faça um programa para perguntar 10 números e informar a soma total destes números.

````python
sum = float(input('Entre com o número: '))

count = 0

while count < 9:
  sum += float(input('Entre com o número: '))

  count += 1

print('Soma: {}'.format(sum))
````

# Questão 2
Faça um programa que escreva todos os números múltiplos de 7 entre 1 e N, sendo N um valor introduzido pelo usuário. Por exemplos: 7, 14, 21, 28, 35.

````python
limit = int(input('Entre com o limite: '))

count = 1

multiples = ''

while count <= limit:
  if count == limit:
    multiples += '{}'.format(count * 7)
  else: 
    multiples += '{}, '.format(count * 7) 

  count += 1

print(multiples)
````

# Questão 4
Faça a multiplicação entre dois números usando somente soma.

````python
number1 = int(input('Numero 1: '))

number2 = int(input('Numero 2: '))

sum = 0

count = 0

while count < number2:
  sum += number1

  count += 1

print('Resultado = {}'.format(sum))
````

# Questão 5
Faça um programa, usando loops, que peça para um usuário digitar um número e que só finaliza quando o usuário digitar 0. Ao final imprima a soma de todos os números digitados.

````python
number = -1

sum = 0

while not number == 0:
  number = float(input('Digite um numero: '))

  sum += number

print('Soma = {}'.format(sum))
````

# Questão 6
Faça um programa que lê uma indeterminada quantidade de números até que o 0 (zero) seja digitado. Ao final, mostre os três menores de forma decrescente.

````python
sum = 0

numbers = []

number = float(input('Digite um numero: '))

while not number == 0:
  if (len(numbers) == 0):
    numbers.insert(0, number)
  else:
    count = 0

    while count < (len(numbers)):
      if numbers[count] > number:
        numbers.insert(count, number)

        count = len(numbers)
  
      count += 1

      if (numbers[-1] != number):
        numbers.append(number)

  while len(numbers) > 3:
    numbers.pop()

  number = float(input('Digite um numero: '))

numbers.sort(reverse=True)

print(numbers)
````

# Questão 7
Informe, de forma decrescente todos os pares entre N (número fornecido pelo usuário) e -N.

````python
number = int(input('Entre com o numero: '))

count = - number

pair_numbers = []

while count <= number:
  if count % 2 == 0:
    pair_numbers.append(count)
  
  count += 1

print(pair_numbers)
````

# Questão 8
Faça um programa para simular uma situação simples de depósito, retirada e consulta em um banco. Exiba o seguinte menu com as opções:
```
1 - Depósito
2 - Retirada
3 - Saldo
4 - Sair do algoritmo
```
Se a escolha do usuário for depósito ou retirada, o algoritmo deverá pedir o valor da operação e atualizar automaticamente o valor existente na conta. O algoritmo deverá ser utilizado até que o usuário escolha a opção sair do algoritmo.

````python
def print_menu():
  print('1 - Depósito')
  print('2 - Retirada')
  print('3 - Saldo')
  print('4 - Sair do algoritmo')

def deposit(balance):
  value = float(input('Valor de deposito: '))

  return (balance + value)

def withdraw(balance):
  value = float(input('Valor de saque: '))

  if (balance < value):
    print('Saldo insuficiente')

    return balance
  else:
    return (balance -  value)

def bank():
  option = -1

  balance = 0

  while option != 4:
    print_menu()

    option = int(input())

    if option == 1:
      balance = deposit(balance)
    elif option == 2:
      balance = withdraw(balance)
    elif option == 3:
      print('Saldo: R$ {}'.format(balance))

bank()
````
