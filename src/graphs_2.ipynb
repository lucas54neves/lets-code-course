````python
class Graph:
    def __init__(self):
        self.adjacency_matrix = {}

    def add_vertex(self, index):
        self.adjacency_matrix[index] = {}
    
    def add_edge(self, vertex1, vertex2):
        self.adjacency_matrix[vertex1][vertex2] = 1
        self.adjacency_matrix[vertex2][vertex1] = 1

    def remove_edge(self, vertex1, vertex2):
        if self.adjacency_matrix.get(vertex1).get(vertex2):
            self.adjacency_matrix[vertex1][vertex2] = 0

    def __repr__(self):
        matrix = '  | '

        for vertex1 in self.adjacency_matrix:
            matrix += str(vertex1) + ' '
        
        matrix += '\n'

        for vertex1 in self.adjacency_matrix:
            matrix += str(vertex1) + ' | '
            for vertex2 in self.adjacency_matrix:
                matrix += str(self.adjacency_matrix[vertex1].get(vertex2, 0)) + ' '
            matrix += '\n'

        return matrix

the_graph = Graph()

the_graph.add_vertex(0)
the_graph.add_vertex(1)
the_graph.add_vertex(2)
the_graph.add_vertex(3)
the_graph.add_vertex(4)
the_graph.add_vertex(5)
the_graph.add_vertex(6)

the_graph.add_edge(1, 3)
the_graph.add_edge(2, 3)
the_graph.add_edge(2, 5)
the_graph.add_edge(3, 1)
the_graph.add_edge(4, 1)
the_graph.add_edge(4, 3)
the_graph.add_edge(4, 6)
the_graph.add_edge(4, 6)
the_graph.add_edge(5, 2)
the_graph.add_edge(5, 4)
the_graph.add_edge(5, 5)
the_graph.add_edge(5, 6)
the_graph.add_edge(6, 4)
the_graph.add_edge(6, 5)

print(the_graph)
````

# Questão 3

Considere um determinado número de cidades (N) e uma determinada quantidade de rotas (M) que você sabe que existem entre aquelas cidades. Com base nisso, você deve criar um script em Python que recebe um número N de cidades e um número M de caminhos que existem entre elas. Em seguida, seu programa vai ler quais são esses M caminhos e a distância deles.

Para exemplificar, vamos analisar a entrada abaixo:

```
3 2  
0 1 10  
1 2 23  
```

O exemplo acima é bem simples. Os dois primeiros números são o N e o M; ou seja, existem 3 cidades e 2 caminhos conhecidos entre elas. Em seguida, temos os dois caminhos:

- O primeiro vai da cidade 0 para a cidade 1, cuja distância é de 10.
- O segundo vai da cidade 1 para a cidade 2, cuja distância é de 23.

Teste o seu programa para a entrada acima e, também, para a entrada abaixo:

```
7 11
0 1 7
0 3 5
1 2 8
1 3 9
1 4 7
2 4 5
3 4 15
3 5 6
4 5 8
4 6 9
5 6 11
```

````python
class Edge:
    def __init__(self, node, weight):
        self.node = node
        self.weight = weight

class Node:
    def __init__(self, index):
        self.index = index
        self._distance = float('inf')
        self._parent = None
        self._neighbors = []
    
    @property
    def distance(self):
        return self._distance
    
    @distance.setter
    def distance(self, new_distance):
        self._distance = new_distance
    
    @property
    def parent(self):
        return self._parent
    
    @parent.setter
    def parent(self, node):
        self._parent = node
    
    @property
    def neighbors(self):
        return self._neighbors
    
    @neighbors.setter
    def neighbors(self, data):
        node, weight = data

        self._neighbors.append(Edge(node, weight))

class Graph:
    def __init__(self):
        self._nodes = {}
    
    @property
    def nodes(self):
        return self._nodes
    
    @nodes.setter
    def nodes(self, index):
        self.nodes[index] = Node(index)
    
    def get_node_by_index(self, index):
        return self.nodes.get(index)

    def add_edge(self, index1, index2, weight):
        node1 = self.get_node_by_index(index1)

        node2 = self.get_node_by_index(index2)

        if node1 and node2:
            node1.neighbors = node2, weight
            node2.neighbors = node1, weight

    @staticmethod
    def extract_shortest_distance_node(node_list):
        node_list.sort(key=lambda node: node.distance)

        shortest_distance_node = node_list[0]

        node_list = node_list[1:]

        return shortest_distance_node, node_list
    
    @staticmethod
    def relax(node, neighbor, weight):
        if neighbor.distance > node.distance + weight:
            neighbor.distance = node.distance + weight

            neighbor.parent = node

    def dijkstra(self, source):
        source_node = self.get_node_by_index(source)

        source_node.distance = 0

        distances = []

        queue = list(self.nodes.values())

        while queue:
            node, queue = self.extract_shortest_distance_node(queue)

            distances.append(node)

            for neighbor in node.neighbors:
                self.relax(node, neighbor.node, neighbor.weight)

        for node in distances:
            print(f'Node: {node.index} | Distance: {node.distance}')

def run():
    graph = Graph()

    data = input()

    data = data.split(' ')

    number_of_nodes = int(data[0])

    number_of_edge = int(data[1])

    for i in range(number_of_nodes):
        graph.nodes = i

    for j in range(number_of_edge):
        data = input()

        data = data.split(' ')

        index1 = int(data[0])

        index2 = int(data[1])

        weight = int(data[2])
        
        graph.add_edge(index1, index2, weight)
    
    graph.dijkstra(0)

run()
````

# Questão 4

Implemente o grafo ponderado ilustrado pela figura abaixo, utilizando as representações baseadas em:

1. matriz de adjacências;
2. lista de adjacências.

Em seguida, exiba a matriz e a lista que representam o grafo.

<img src="https://ucarecdn.com/a67cb888-aa0c-424b-8c7f-847e38dd5691/" alt="drawing" width="500"/>

````python
class Graph:
    def __init__(self):
        self.adjacency_matrix = {}
        self.adjacency_list = {}
    
    def add_vertex(self, index):
        self.adjacency_matrix[index] = {}
        self.adjacency_list[index] = []
    
    def add_edge(self, index_1, index_2, weight):
        self.adjacency_matrix[index_1][index_2] = weight
        self.adjacency_matrix[index_2][index_1] = weight
        
        self.adjacency_list[index_1].append((index_2, weight))
        self.adjacency_list[index_2].append((index_1, weight))
    
    def show_adjacency_matrix(self):
        representation = '   |'

        for i in self.adjacency_matrix:
            representation += f' {i}'
        
        representation += '\n'

        for i in self.adjacency_matrix:
            representation += f' {i} |'

            for j in self.adjacency_matrix:
                representation += f' {0 if not (self.adjacency_matrix.get(i) and self.adjacency_matrix.get(i).get(j)) else self.adjacency_matrix.get(i).get(j)}'
            
            representation += '\n'
        
        return representation
    
    def show_adjacency_list(self):
        representation = ''

        for i in self.adjacency_list:
            representation += f'{i}: {self.adjacency_list.get(i)}\n'

        return representation

graph = Graph()

graph.add_vertex(0)
graph.add_vertex(1)
graph.add_vertex(2)
graph.add_vertex(3)
graph.add_vertex(4)

graph.add_edge(0, 1, 3)
graph.add_edge(0, 3, 7)
graph.add_edge(0, 4, 8)
graph.add_edge(1, 2, 1)
graph.add_edge(1, 3, 4)
graph.add_edge(2, 3, 2)
graph.add_edge(3, 4, 3)

print(graph.show_adjacency_matrix())
print(graph.show_adjacency_list())
````
