**Exercício 1**

Calcule a média de um aluno com os seguintes critérios:

Média da provas: MP = (P1 + 2 P2 + 2 P3)/5

Média dos exercícios: ME = (E1 + E2 + 2 E3 + 3 E4)/7

Média final: se	MP e ME são maiores ou iguais a 5, então MF = (2 MP + ME)/3. Senão MF = mínimo entre MP e ME

```python
tests = []

exercises = []

count = 0

while count < 3:
  grade = float(input('Digite a nota da prova: '))

  tests.append(grade)

  count += 1

count = 0

while count < 4:
  grade = float(input('Digite a nota do exercicio: '))

  exercises.append(grade)

  count += 1

tests_average = (tests[0] + 2 * tests[1] + 2 * tests[2]) / 5

exercises_average = (exercises[0] + exercises[1] + 2 * exercises[2] + 3 * exercises[3]) / 7

final_average = 0

if tests_average >= 5 and exercises_average >= 5:
  final_average = (2 * tests_average + exercises_average) / 3
elif exercises_average < tests_average:
  final_average = exercises_average
else:
  final_average = tests_average

print('Media final: {:.2f}'.format(final_average))
```

Ajuste o resultado anterior para calcular as médias de vários alunos

**Exercício 3**

Dados dois vetores x e y, ambos com n elementos, determinar o produto escalar desses vetores.

Produto escalar:

[1, 2] x [3, 4] = 1\*3 + 2\*4 = 11
 

```python
vector_size = int(input('Digite o n: '))

vector1 = []

vector2 = []

product = 0

count = 0

while count < vector_size:
  element = int(input('Digite um elemento para o primeiro vetor: '))

  vector1.append(element)

  count += 1

count = 0

while count < vector_size:
  element = int(input('Digite um elemento para o segundo vetor: '))

  vector2.append(element)

  count += 1

for i in range(vector_size):
  product += vector1[i] * vector2[i]

print('Produto escalar: {}'.format(product))
```

**Exercício 4**

Dadas duas listas de números inteiros, mostre todos os possíveis pares entre eles.

Exemplo:

lista 1 = [1, 2]

lista 2 = [1, 3, 5]

pares = (1, 1), (2, 1), (1, 3), (2, 3), (1, 5), (2, 5)

```python
list1 = [1, 2]

list2 = [1, 3, 5]

pairs = []

for i in range(len(list1)):
  for j in range(len(list2)):
    pairs.append((list1[i], list2[j]))

print(pairs)
```
