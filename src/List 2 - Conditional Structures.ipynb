# Questão 1

Escreva um programa que peça a nota de 3 provas de um aluno e verifique se ele passou ou não de ano.

Obs.: O aluno irá passar de ano se sua média for maior que 6.

```python
grade1 = float(input('Nota 1: '))

grade2 = float(input('Nota 2: '))

grade3 = float(input('Nota 3: '))

average = (grade1 + grade2 + grade3) / 3

if average > 6:
  print('Aprovado')
else:
  print('Reprovado')
```

# Questão 2

Desafio 1 - Faça um programa que leia 3 números e informe o maior deles.

```python
number1 = float(input('Entre com o primeiro numero: '))

number2 = float(input('Entre com o segundo numero: '))

number3 = float(input('Entre com o terceiro numero: '))

biggest = number1

if biggest < number2:
  biggest = number2

if biggest < number3:
  biggest = number3

print('Maior = {}'.format(biggest))
```

# Questão 3

Desafio 2 - Faça o mesmo programa do exercício anterior, porém com 4 números.


```python
number1 = float(input('Entre com o primeiro numero: '))

number2 = float(input('Entre com o segundo numero: '))

number3 = float(input('Entre com o terceiro numero: '))

number4 = float(input('Entre com o quarto numero: '))

biggest = number1

if biggest < number2:
  biggest = number2

if biggest < number3:
  biggest = number3

if biggest < number4:
  biggest = number4

print('Maior = {}'.format(biggest))
```

# Questão 4

Desafio 3 - Um hospital quer fazer o diagnóstico de gripe ou dengue a partir de um questionário de sintomas, tendo as perguntas abaixo, faça um programa que faça o diagnóstico deste hospital:

a. Sente dor no corpo?

b. Você tem febre?

c. Você tem tosse?

d. Está com congestão nasal?

e. Tem manchas pelo corpo?

Para o diagnóstico ele tem a seguinte tabela:

A | B	| C	| D	| E | Resultado
:-: | :-: | :-: | :-: | :-: |:-:
Sim	| Sim	| Não	| Não	| Sim | Dengue
Sim	| Sim	| Sim	| Sim	| Não | Gripe
Não	| Sim	| Sim	| Sim	| Não	| Gripe
Sim	| Não	| Não	| Não	| Não	| Sem doenças
Não	| Não	| Não	| Não	| Não	| Sem doenças

```python
answer1 = input('Sente dor no corpo? [S/N] ')

answer2 = input('Você tem febre? [S/N] ')

answer3 = input('Você tem tosse? [S/N] ')

answer4 = input('Está com congestão nasal? [S/N] ')

answer5 = input('Tem manchas pelo corpo? [S/N] ')

if answer1 == 'S' and answer2 == 'S' and answer3 == 'N' and answer4 == 'N' and answer5 == 'S':
  print('Dengue')
elif (answer1 == 'S' or answer1 == 'N') and answer2 == 'S' and answer3 == 'S' and answer4 == 'S' and answer5 == 'N':
  print('Gripe')
elif (answer1 == 'S' or answer1 == 'N') and answer2 == 'N' and answer3 == 'N' and answer4 == 'N' and answer5 == 'N':
  print('Sem doenças')
else:
  print('Causa desconhecida')
```
