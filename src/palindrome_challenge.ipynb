# Desafio do Palíndromo
Crie uma função recursiva que verifique se uma palavra é um palindromo. Depois, o objetivo de vocês é dado o csv: ```palindromo_exemplo.csv```, iterar nas frases (palindromas) e retornar __True__ para todas as frases do CSV.

````python
import csv
import re

def is_palindrome(word):
    if len(word) == 1:
        return True
    
    if len(word) == 2:
        return word[0] == word[1]

    return word[0] == word[-1] and is_palindrome(word[1:-1])

def preprocessing():
    with open('../doc/palindromo_exemplo.csv') as file_csv:
        reader = csv.reader(file_csv, delimiter='\n')

        for row in reader:
            word = row[0].lower()
            word = word.replace(' ', '')
            word = word.replace('…', '')
            word = re.sub(r'[áàãâä]', 'a', word)
            word = re.sub(r'[éêẽèë]', 'e', word)
            word = re.sub(r'[íìïĩî]', 'i', word)
            word = re.sub(r'[óòõôö]', 'o', word)
            word = re.sub(r'[úùûũü]', 'u', word)
            word = re.sub(r'[ç]', 'c', word)
            word = re.sub(r'[?!,.:;-]', '', word)
            word = re.sub(r'[\'"]', '', word)
            word = re.sub(r'[\“\”]', '', word)
            
            print(f'{word}: {is_palindrome(word)}')

preprocessing()
````
