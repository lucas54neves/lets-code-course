# Questão 1
No código abaixo, estamos interagindo com duas listas que se complementam:
```
lista_lanches = ['x-salada', 'x-burger', 'x-bacon']
lista_precos = [10, 9, 11]
i = 0
while i < 3:
    print(lista_lanches[i], 'custa R$', lista_precos[i])
    i = i + 1
```
Sobre esse tipo de execução:

````python
lista_lanches = ['x-salada', 'x-burger', 'x-bacon']
lista_precos = [10, 9, 11]
i = 0
while i < 3:
    print(lista_lanches[i], 'custa R$', lista_precos[i])
    i = i + 1
````
