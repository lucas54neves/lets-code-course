```python
'''
Faça um programa que peça a temperatura em graus Fahrenheit (F), transforme e mostre a temperatura em graus Celsius (°C).

°C = (5 * (F-32) / 9)

Obs: Tente também fazer um programa que faça o inverso: peça a temperatura em graus Celsius e a transforme em graus Fahrenheit.
'''
fahrenheit_to_celcius = float(input('Entre com a temperatura em Fahrenheit: '))

celcius = 5 * (fahrenheit_to_celcius - 32) / 9

print('A temperatura em Celcius e {:.2f}° C'.format(celcius))

celcius_to_fahrenheit = float(input('Entre com a temperatura em Celcius: '))

fahrenheit = 9 / 5 * celcius_to_fahrenheit + 32

print('A temperatura em Fahrenheit e {:.2f}° F'.format(fahrenheit))
```

```python
# Faça um programa que peça o dia, o mês e o ano para o usuário e imprima a data completa no formato dd/mm/aaaa.
day = int(input('Dia: '))

month = int(input('Mes: '))

year = input('Ano: ')

day = '{}'.format(day) if day >= 10 else '0{}'.format(day)

month = '{}'.format(month) if month >= 10 else '0{}'.format(month)

date = '{}/{}/{}'.format(day, month, year)

print(date)
```

```python
'''
Faça um programa que peça 2 números inteiros e um número real, calcule e mostre:

a) o produto do dobro do primeiro com metade do segundo.

b) a soma do triplo do primeiro com o terceiro.

c) o terceiro elevado ao cubo.
'''
number1 = int(input('Primeiro numero inteiro: '))

number2 = int(input('Segundo numero inteiro: '))

number3 = float(input('Numero real: '))

first = number1

if number2 > first:
  second = first
  
  first = number2
else:
  second = number2

if number3 > first:
  third = second

  second = first

  first = number3
elif number3 > second:
  third = second

  second = number3
else:
  third = number3

result1 = (first * 2) * (second / 2)

result2 = (first * 3) + (third)

result3 = third ** 3

print('a) {:.2f}'.format(result1))

print('b) {:.2f}'.format(result2))

print('c) {:.2f}'.format(result3))
```

```python
'''
  Faça um programa que peça o peso e altura de uma pessoa e calcule seu IMC (Índice de Massa Corporal).

  Obs: IMC = Peso/Altura**2
'''
weight = float(input('Peso: '))

height = float(input('Altura: '))

index = weight / (height ** 2)

if index < 18.5:
  message = 'IMC: {:.2f} - Classificacao: Magreza - Obesidade: 0'.format(index)
elif index >= 18.5 and index < 25:
  message = 'IMC: {:.2f} - Classificacao: Normal - Obesidade: 0'.format(index)
elif index >= 25 and index < 30:
  message = 'IMC: {:.2f} - Classificacao: Sobrepeso - Obesidade: I'.format(index)
elif index >= 30 and index < 40:
  message = 'IMC: {:.2f} - Classificacao: Obesidade - Obesidade: II'.format(index)
else:
  message = 'IMC: {:.2f} - Classificacao: Obesidade grave - Obesidade: III'.format(index)

print(message)
```

```python
'''
Desafio 1 - Peça para o usuário digitar uma velocidade inicial (em m/s), uma posição inicial (em m) e um instante de tempo (em s) e imprima a posição de um projétil nesse instante de tempo.

Use a fórmula matemática:

y(t) = y(0) + v(0)*t + (g*(t**2)/2)

Onde, g é a aceleração da gravidade (-10m/s²), y(t) é a posição final, y(0) é a posição inicial, v(0) é a velocidade inicial e t é o instante de tempo.
'''
initial_velocity = float(input('Entre com a velocidade inicial: '))

starting_position = float(input('Entre com a posicao inicial: '))

instant_of_time = float(input('Entre com um instante de tempo: '))

position = starting_position + initial_velocity * instant_of_time - (10 * (instant_of_time ** 2) / 2)

print('O posicao do projetil nesse instante de tempo e {}'.format(position))
```
